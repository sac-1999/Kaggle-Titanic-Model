# Data Preprocessing

```python
import pandas as pd
import seaborn as sns
import numpy as np


data = pd.read_csv('train.csv')
data['Cabin'].value_counts().shape





data['Cabin']=np.where(data['Cabin'].isnull() , 0 ,1)
data['Embarked']=np.where(data['Embarked'].isnull() , data['Embarked'].mode() ,data['Embarked'])
data['Age']=np.where(data['Age'].isnull() , data['Age'].median() ,data['Age'])
data['Ticket_cat']=data['Ticket'].astype('category')
data['Ticket_cat'] = data['Ticket_cat'].cat.codes
data['Ticket_cat'].value_counts()
data['Ticket'] = data['Ticket_cat']

pd.set_option('display.max_rows',len(data))

#Name Encoding
respect = []
for names in data[['Name']].values.tolist():
    respect.append((names[0].split(',')[1].split('.')[0]).strip())
data['Name']=pd.DataFrame(respect)

data['Name_cat'] = data['Name'].astype('category')
data['Name_cat'] = data['Name_cat'].cat.codes
data['Name'] = data['Name_cat']
data


x = data.dtypes
x = x.to_frame()
x = x[x[0]==object]
x = x.index.tolist()

dummies = pd.get_dummies(data[x],drop_first='True')
dummies

data = pd.concat([data,dummies],axis=1).drop(x,axis=1)
data.isnull().sum()
data.shape
data

data.drop(labels=['PassengerId','Ticket_cat','Name_cat'],axis=1,inplace=True)

data.shape

correlation = data.corr()
correlation[correlation<-0.5]



x = data.iloc[:,1:]
y = data.iloc[:,:1]



from sklearn.ensemble import RandomForestClassifier  
random_classifier= RandomForestClassifier(n_estimators= 14, criterion="gini")  
random_classifier.fit(x, y)  

random_pred_train = random_classifier.predict(x)
random_pred_train = random_pred_train.reshape((-1,1))

from sklearn import metrics
metrics.confusion_matrix(y,random_pred_train)

metrics.accuracy_score(y,random_pred_train)


            
```

```python
import pandas as pd
import seaborn as sns
import numpy as np


data = pd.read_csv('train.csv')

#sns.heatmap(data.isnull())
data_copy_1 = data
data.isnull().sum()


#drop cabin column 
data = data.drop('Cabin',axis=1)

#drop two rows for embarked missing
#data = data.drop (data[data['Embarked'].isnull()])
drop_rows = list(data[data['Embarked'].isnull()].index)
data = data.drop(drop_rows)

data.isnull().sum()
mean_age = data['Age'].mean()
data['Age'] = np.where(data['Age'].isnull(),mean_age,data['Age'])


#droping all useless colns before applying one hot encoding in data
data = data.drop(['PassengerId','Name','Ticket'],axis=1)

data.shape
drop_rows
data = data.reset_index()
data.drop(labels = 'index',axis=1,inplace = True)
```

```python
#Implementing one hot encoding 
x = data.dtypes
x = x.to_frame()
x = x[x[0]==object]
x = x.index.tolist()

dummies = pd.get_dummies(data[x],drop_first='True')
dummies

data = pd.concat([data,dummies],axis=1).drop(x,axis=1)
data.isnull().sum()
data.shape
data
```

```python
from sklearn.preprocessing import MinMaxScaler

data_scaled = data.copy()


#trans_t = ('min_max', MinMaxScaler(), [2,5])
from sklearn.compose import ColumnTransformer
coln_trans = ColumnTransformer(transformers = [('min_max', MinMaxScaler(),[2,5])], remainder= 'passthrough')
age_fare = pd.DataFrame(coln_trans.fit_transform(data_scaled))
data_scaled['Age']=age_fare[0]
data_scaled['Fare']=age_fare[1]

x = data_scaled.iloc[:,1:]
y = data_scaled.iloc[:,:1]


data_scaled
```

```python
data_scaled
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
x_train , x_test ,y_train,y_test = train_test_split(x,y,test_size=1)

x_test.shape,y_test.shape
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logistic_classifier = LogisticRegression()
logistic_classifier.fit(x_train,y_train)
logistic_pred_test = logistic_classifier.predict(x_test)
logistic_pred_test = logistic_pred_test.reshape((-1,1))

logistic_pred_test.shape,y_test.shape
#Training data accuracy 
from sklearn import metrics
print(metrics.accuracy_score(y_test,logistic_pred_test))

metrics.confusion_matrix(y_test,logistic_pred_test)
```

# Linear Regresion Model

```python
from sklearn.linear_model import LinearRegression
linear_classifier = LinearRegression()
linear_classifier.fit(x_train,y_train)
linear_pred_train = linear_classifier.predict(x_train)
linear_pred_train = linear_pred_train.reshape((-1,1))

linear_data = pd.DataFrame(linear_pred_train)
linear_pred_train = np.where(linear_data[0]>=0.5,1,0)
linear_pred_train
print(metrics.confusion_matrix(y_train,linear_pred_train))

metrics.accuracy_score(y_train,linear_pred_train)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier  
random_classifier= RandomForestClassifier(n_estimators= 14, criterion="gini")  
random_classifier.fit(x_train, y_train)  

random_pred_train = random_classifier.predict(x_test)
random_pred_train = random_pred_train.reshape((-1,1))

from sklearn import metrics
metrics.confusion_matrix(y_test,random_pred_train)

metrics.accuracy_score(y_test,random_pred_train)
        
        
```
